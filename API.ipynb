```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
df = pd.DataFrame()
i = 1
while True:
    print(i)
    r = requests.get("https://www.climatewatchdata.org/api/v1/data/historical_emissions",params={"page": i})
    if not r.json()["data"]:
        break
    
    df = pd.concat([df, pd.DataFrame(r.json()["data"])], ignore_index=True)
    print(r.json()["data"][0]["iso_code3"])
    i += 1
    
```

```python
df.to_excel("data.xlsx",index=False)
```

```python
df = pd.read_excel("data.xlsx")
```

```python
df["emissions"] = df["emissions"].apply(lambda x: json.loads(x.replace("'", "\"").replace('None', 'null')))
```

```python
df = df.explode(column="emissions", ignore_index=True)
df.head()
```

```python
df[["year","emission"]] = df['emissions'].apply(pd.Series)
```

```python
df = df.drop(columns=["emissions"])
```

```python
df["year"] = df["year"].astype(int)
```

```python
df.to_csv("data.csv",index=False)
```
